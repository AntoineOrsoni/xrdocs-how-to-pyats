```python
from genie import testbed
from unicon.eal.dialogs import Statement, Dialog
import xmltodict

testbed = testbed.load('./testbed.yaml')
source = testbed.devices['xrd-source']

source.connect(via='netconf', log_stdout=True)
source.connect(via='cli', log_stdout=False)
```

```python
rpc_request = '''
<rpc xmlns="urn:ietf:params:xml:ns:netconf:base:1.0" message-id="101">
    <traceroute xmlns="http://cisco.com/ns/yang/Cisco-IOS-XR-traceroute-act">
        <destination>
        <destination>10.3.1.3</destination>
        </destination>
    </traceroute>
</rpc>
'''

reply_netconf = source.netconf.request(rpc_request, timeout=120)
reply_cli = source.cli.execute('traceroute 10.3.1.3')
```

```python
print(reply_cli)
print(reply_netconf)
```
